```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter

# Data
questions = ['A', 'B', 'C', 'D', 'E']
labels = ['Strongly agree', 'Agree', 'Neither agree nor disagree', 'Disagree', 'Strongly disagree']
data = {
    'A': [11, 10, 2, 1, 0],
    'B': [6, 14, 3, 1, 0],
    'C': [9, 14, 1, 0, 0],
    'D': [10, 5, 7, 2, 0],
    'E': [10, 11, 2, 1, 0]
}

# Convert raw numbers to percentage
total_per_question = {question: sum(values) for question, values in data.items()}
data_percentage = {question: [value/total_per_question[question]*100 for value in values] for question, values in data.items()}

# Plot
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each category
bottoms = np.zeros(len(questions))
for idx, label in enumerate(labels):
    values = [data_percentage[question][idx] for question in questions]
    ax.bar(questions, values, label=label, bottom=bottoms,alpha=0.8)
    bottoms += values
plt.rcParams.update({'font.size': 14})
# Settings and show
ax.set_ylabel('Percentage (%)')
# ax.set_title('Likert Questions Results in Percentage')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.09), ncol=len(labels))
ax.yaxis.set_major_formatter(PercentFormatter(100))  # Format y-axis to show percentages

plt.rcParams.update({'font.size': 16})
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter


# Data
ccg_codes = ['E38000243', 'E38000136', 'E38000242', 'E38000244']
with_river = [1, 0.418, 0.79, 0.256]
without_river = [0.97, 0.4, 0.3, 0.13]


bar_width = 0.35
index = np.arange(len(ccg_codes))

fig, ax = plt.subplots(figsize=(12, 8))

bar1 = ax.bar(index, with_river, bar_width, label='With River', yerr=[0,0.11,0.07,0.07], alpha=0.8, capsize=10)
bar2 = ax.bar(index + bar_width, without_river, bar_width, label='Without River', yerr=[0.03,0.12,0.08,0.07], alpha=0.8, capsize=10)

ax.set_xlabel('Target CCG\'s Code')
ax.set_ylabel('Average Accuracy (%)')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(ccg_codes, fontsize=14)
ax.legend(fontsize=14)

# Format y-axis to show percentages and increase font size
ax.yaxis.set_major_formatter(PercentFormatter(1))
ax.tick_params(axis='y', labelsize=14)

plt.rcParams.update({'font.size': 16})
plt.tight_layout()
plt.show()
```
